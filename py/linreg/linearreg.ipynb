```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# Base function and noise
noise = 0.3
size = 100
def func(X):
    return 2.7 * X + 2
```

```python
# Sampling
np.random.seed(2804)


X = np.random.uniform(size=size, high = 2.0)
epsilon = np.random.normal(size=size, scale = noise)
y = func(X) + epsilon

# We have no abline in matplotlib 
# https://github.com/matplotlib/matplotlib/issues/5253
xrange = np.array([min(X), max(X)])
```

```python
# Drawing scatter and original line
plt.figure(0)
plt.scatter(X, y,  color='black', s=.5)
plt.plot(xrange, func(xrange), color='blue', linewidth=0.8, linestyle="--", dashes=(2, 4))
```

```python
# Fitting regression
from sklearn import linear_model
regr = linear_model.LinearRegression()
# there are problems with one dimensional feature
regr.fit(X.reshape(-1,1), y)

print(regr.coef_, regr.intercept_)
```

```python
# Drawing all togetger
plt.figure(1)
plt.scatter(X, y,  color='black', s=.5)
plt.plot(xrange, func(xrange), color='blue', linewidth=0.8, linestyle="--", dashes=(2, 4))
y_predicted = regr.predict(xrange.reshape(-1,1))
plt.plot(xrange, y_predicted, color='red', linewidth=0.5)
```
